```python
# Install sylearn dan library pendukung
!pip install openpyxl pandas scikit-learn matplotlib

import pandas as pd
# Change the import statement to use sklearn
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from google.colab import files
import matplotlib.pyplot as plt

# Upload file iris.xlsx
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Baca data
df = pd.read_excel(filename)

# Hapus kolom 'Class' dan 'id' (jika ada)
X = df.drop(columns=['Class', 'id'], errors='ignore')

sse_list = []
silhouette_list = []
klaster_range = range(2, 5)

for k in klaster_range:
    # KMeans is imported from sklearn, no need to change here
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    labels = kmeans.labels_
    sse = kmeans.inertia_
    silhouette = silhouette_score(X, labels)
    sse_list.append(sse)
    silhouette_list.append(silhouette)
    print(f'k={k}, SSE={sse:.3f}, Silhouette={silhouette:.3f}')

# Tentukan k terbaik berdasarkan silhouette score
best_k_index = silhouette_list.index(max(silhouette_list))
best_k = klaster_range[best_k_index]
print(f'\nKlaster terbaik berdasarkan silhouette score: k={best_k}')

# Visualisasi SSE dan Silhouette Score
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(list(klaster_range), sse_list, marker='o')
plt.title('SSE vs Jumlah Klaster')
plt.xlabel('Jumlah Klaster (k)')
plt.ylabel('SSE')

plt.subplot(1,2,2)
plt.plot(list(klaster_range), silhouette_list, marker='o', color='orange')
plt.title('Silhouette Score vs Jumlah Klaster')
plt.xlabel('Jumlah Klaster (k)')
plt.ylabel('Silhouette Score')

plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Lakukan PCA 2 dimensi untuk visualisasi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Lakukan K-Means dengan k terbaik
from sklearn.cluster import KMeans
kmeans_best = KMeans(n_clusters=best_k, random_state=42, n_init=10)
kmeans_best.fit(X)
labels_best = kmeans_best.labels_

# Plot scatter PCA hasil klastering
plt.figure(figsize=(8,6))
colors = ['red', 'blue', 'green', 'orange']

for i in range(best_k):
    cluster_points = X_pca[labels_best == i]
    plt.scatter(cluster_points[:,0], cluster_points[:,1], label=f'Cluster {i}', alpha=0.7, color=colors[i])

plt.title(f'Scatter Plot Hasil K-Means Clustering (k={best_k})')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Tampilkan nilai evaluasi dalam bentuk DataFrame
result = pd.DataFrame({
    'Jumlah Klaster (k)': list(klaster_range), # Ganti k_values dengan klaster_range
    'SSE': sse_list, # Ganti sse dengan sse_list
    'Silhouette Score': silhouette_list # Ganti silhouette_scores dengan silhouette_list
})

print("Hasil Evaluasi Clustering:")
print(result)
```

```python
# Install library yang dibutuhkan
!pip install scikit-learn openpyxl pandas matplotlib scipy

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score
from scipy.optimize import linear_sum_assignment
from sklearn.preprocessing import LabelEncoder
from google.colab import files

# Upload file iris.xlsx
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Baca dan siapkan data
df = pd.read_excel(filename)
X = df.drop(columns=['id', 'Class'], errors='ignore')
y_true = df['Class']

# Encode label Class (misal: setosa=0, versicolor=1, virginica=2)
le = LabelEncoder()
y_encoded = le.fit_transform(y_true)

# Clustering KMeans
k = 3  # karena iris memiliki 3 kelas
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Confusion Matrix
cm = confusion_matrix(y_encoded, y_kmeans)

# Cari label mapping terbaik dengan Hungarian Algorithm
row_ind, col_ind = linear_sum_assignment(-cm)  # Maksimalkan akurasi
optimal_mapping = dict(zip(col_ind, row_ind))

# Terapkan mapping label ke hasil clustering
y_kmeans_mapped = [optimal_mapping[label] for label in y_kmeans]

# Hitung akurasi
akurasi = accuracy_score(y_encoded, y_kmeans_mapped)
print(f'Akurasi Clustering terhadap label asli: {akurasi*100:.2f}%')

# Tampilkan confusion matrix
print("\nConfusion Matrix (Setelah Mapping):")
print(confusion_matrix(y_encoded, y_kmeans_mapped))
```
