**DATA UNDERSTANDING**

```python
# Import library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("/content/drive/MyDrive/dataUTS/cirrhosis.csv")

# 1. Informasi umum
print("ðŸ”¹ Informasi Dataset:")
print(df.info())

# 2. Beberapa baris awal
print("\nðŸ”¹ Contoh Data:")
display(df.head())

# 3. Statistik deskriptif
print("\nðŸ”¹ Statistik Deskriptif:")
display(df.describe(include='all'))

# 4. Jumlah missing values
print("\nðŸ”¹ Missing Values Tiap Kolom:")
print(df.isnull().sum())

# 5. Distribusi target (Status)
print("\nðŸ”¹ Distribusi Kelas Target (Status):")
print(df['Status'].value_counts())

sns.countplot(data=df, x='Status')
plt.title("Distribusi Kelas Status")
plt.show()

# 6. Visualisasi distribusi fitur numerik
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

num_cols = 4
num_rows = int(len(numerical_cols) / num_cols) + (len(numerical_cols) % num_cols > 0)

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.histplot(df[col], kde=True, bins=30, color='teal')
    plt.title(f'Distribusi {col}')
    plt.tight_layout()

plt.suptitle("Distribusi Fitur Numerik", y=1.02, fontsize=16)
plt.show()
```

PREPOCESSING

```python
# Import library yang dibutuhkan
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/dataUTS/cirrhosis.csv")

# 1. Hapus kolom ID (bukan fitur yang berguna)
df.drop(columns=["ID"], inplace=True)

# 2. Imputasi nilai kosong
# Imputasi fitur numerik dengan median
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
imputer_num = SimpleImputer(strategy='median')
df[num_cols] = imputer_num.fit_transform(df[num_cols])

# Imputasi fitur kategorikal dengan modus (frekuensi terbanyak)
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

# 3. Label encoding untuk fitur kategorikal
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# 4. Pisahkan fitur dan label
X = df.drop(columns=["Status"])  # fitur
y = df["Status"]                # label

# 5. Normalisasi/Standardisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 6. Bagi data menjadi train dan test set
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("Preprocessing selesai. Data siap untuk modeling.")
```

**MODELING**

```python
# Import library untuk modeling
from sklearn.ensemble import RandomForestClassifier

# Inisialisasi dan training model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Prediksi pada data uji
y_pred = model.predict(X_test)

print("Model Random Forest telah dilatih dan prediksi selesai.")
```

**EVALUASI**

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
report = classification_report(y_test, y_pred)
print("ðŸ“„ Classification Report:\n")
print(report)

# Accuracy
acc = accuracy_score(y_test, y_pred)
print(f"ðŸŽ¯ Accuracy: {acc:.2f}")
```
